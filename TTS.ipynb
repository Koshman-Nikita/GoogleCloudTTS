```python
from gtts import gTTS

text_to_say = "An elite athlete has just emerged from spending a record-breaking 500 days alone in a cave in Spain. Beatriz Flamini, 50, is an extreme mountaineer."
language = "en"

gtts_object = gTTS(text = text_to_say, 
                  lang = language,
                  slow = False)

gtts_object.save("/content/gtts.wav")
     
```

```python
from IPython.display import Audio

Audio("/content/gtts.wav")
     
```
